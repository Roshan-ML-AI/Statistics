## Q4. Answer the following questions with statistical evidence

## Q4 a. Do charges of people who smoke differ significantly from the people who don't?

```python
sns.boxplot(insurance['smoker'],insurance['charges']);
```

#### <font color=DARKBLUE> From box plot we can see that the distribution of Smokers and non smokers are quite different from each other 
#### <font color=DARKBLUE> We will be performing t-test to understand the if statistically it is significantly different

```python
# T-test to check dependency of smoking on charges
Ho = "Charges of smoker and non-smoker are same"   # Null Hypothesis
Ha = "Charges of smoker and non-smoker are not same"   # Alternate Hypothesis

x = np.array(insurance[insurance.smoker == 'yes'].charges)  #  charges for smokers 
y = np.array(insurance[insurance.smoker == 'no'].charges) #  charges for non-smokers

t, p_value  = stats.ttest_ind(x,y, axis = 0)  #Performing an Independent t-test

print("p_value =",p_value)

# Considering significance level of 5%
    # if the p_value is less than 5% then we reject the null Hypothesis else will accept the null hypothesis

if p_value <= 0.05: 
    print('The charges of people who smoke differ significantly from the people who dont')
else:
    print('Charges of smoker and non-smoker are same')
```

### <font color=red> The charges of people who smoke differ significantly from the people who don't smoke


## Q4 b. Does bmi of males differ significantly from that of females

```python
sns.boxplot(insurance['sex'],insurance['bmi']);
```

#### <font color=DARKBLUE> From box plot we can't see much difference in  distribution of bmi in male and female  
#### <font color=DARKBLUE> We will be performing t-test to understand the if statistically it is significantly different

```python
# T-test to check dependency of bmi on gender
Ho = "Gender has no effect on BMI"   # Stating the Null Hypothesis
Ha = "Gender has an effect on BMI"   # Stating the Alternate Hypothesis

x = np.array(insurance[insurance.sex == 'male'].bmi)  # Selecting bmi values corresponding to males as an array
y = np.array(insurance[insurance.sex == 'female'].bmi) # Selecting bmi values corresponding to females as an array

t, p_value  = stats.ttest_ind(x,y, axis = 0)  #Performing an Independent t-test

print('p value =', round(p_value,3))

# Considering significance level of 5%
    # if the p_value is less than 5% then we reject the null Hypothesis else will accept the null hypothesis

if p_value <= 0.05: 
    print('BMI of males differ significantly from that of females')
else:
    print('Gender has no effect on BMI')
```

### <font color=red> Gender has no effect on BMI

## Q4 c. Is the proportion of smokers significantly different in different genders

```python
print(insurance.sex.value_counts())
sns.countplot(insurance['sex'], hue = insurance['smoker']);
```

#### <font color=DARKBLUE> The non smoker in male is less than the non smoker in females but the smoker in males is higher than the  smokers in females
#### <font color=DARKBLUE> Thus from the plot it can be seen that the proportion of smoker is higher in males

#### <font color=DARKBLUE> We will be performing the proportion z test to check if the proportion difference is statistically significant

```python
# Calculating % share of smokers in male and female
female_smokers = insurance[insurance['sex'] == 'female'].smoker.value_counts()[1]  # number of female smokers
male_smokers = insurance[insurance['sex'] == 'male'].smoker.value_counts()[1] # number of male smokers
females_total = insurance.sex.value_counts()[1] # number of females in the data
males_total = insurance.sex.value_counts()[0] #number of males in the data
```

```python
print([female_smokers, male_smokers] , [females_total, males_total])
print(' Proportion of smokers in females is', round(female_smokers/females_total,2)*100,"% and males =", round(male_smokers/females_total,2)*100,'% ')
```

```python
Ho= "The proportion of smokers is equal for males and females"
Ha = "The proporton of smokers is different in males and females"

from statsmodels.stats.proportion import proportions_ztest

stat, p_value = proportions_ztest([female_smokers, male_smokers] , [females_total, males_total])

print('p value =', round(p_value,3))

 # Considering significance level of 5%
    # if the p_value is less than 5% then we reject the null Hypothesis else will accept the null hypothesis

if p_value <= 0.05:
    print('The proportion of smokers is significantly different in different genders')
else:
    print('The proportion of smokers is equal for males and females')
```

###  <font color=red>The proportion of smokers is significantly different in different genders

## Q4 d. Is the distribution of bmi across women with no children, one child and two children, the same

```python
# Creating a new data frame only with women who has no children, one child or two children

insurance_female=insurance[(insurance['sex'] == 'female')&(insurance['children']<=2)]

# Checking if the data is correctly filtered
print(insurance_female.children.unique())
print(insurance_female.sex.unique())
```

```python
# Visually understanding the distribution of bmi for women by the count of their children
sns.violinplot(insurance_female['children'],insurance_female['bmi']);
```

#### <font color=DARKBLUE> From the Violin plot we can see that the bmi across women with no children, one child and two children are almost similar
#### <font color=DARKBLUE> To check if it is statistically significant we will be performing ANOVA test

```python
Ho = "No. of children has no effect on bmi"   # Null Hypothesis
Ha = "No. of children has an effect on bmi"   # Alternate Hypothesis

zero = insurance_female[insurance_female.children == 0]['bmi']
one = insurance_female[insurance_female.children == 1]['bmi']
two = insurance_female[insurance_female.children == 2]['bmi']

f_stat, p_value = stats.f_oneway(zero,one,two)

print('p value =', round(p_value,3))

 # Considering significance level of 5%
# if the p_value is less than 5% then we reject the null Hypothesis else will accept the null hypothesis

if p_value <= 0.05:  
    print('The distribution of bmi across women with no children, one child and two children is different')
else:
    print('The distribution of bmi across women with no children, one child and two children is same')
```

###  <font color=red> The distribution of BMI across women with no children, one child and two children is same
